```python
samples = loadMCSamples("./chains/union", settings={'smooth_scale_2D':0.9})
samples.updateSettings({'contours': [0.683, 0.957, 0.997]})
```

```python
g = plots.get_subplot_plotter()
g.triangle_plot([samples], ['omegam', 'omega_de'], filled=True)
```

```python
g = plots.get_subplot_plotter()
g.triangle_plot([samples], ['omegam', 'omegak'], filled=True)
```
